```python
###부산진구
import pandas as pd
minwondf = pd.read_csv('부산진구 종합 크롤링 인코딩.csv', encoding='ansi')

minlist = minwondf['title'].to_list()

stopword = ['리','관','롯데','캐슬','포','박형준','건','소음','부','번','공사','앞','및','관련','가야','부암','번길',
        '부산','서면','부산진구','도로','도심','사업','양정','현장','지역','피해','주민','곳','안전','길','옆','부산시','대구',
          '사장','로','처리','가야동','좀','주변','선정','대한','인도','요청','단속','불법','민원','문제','개발','부탁','전포',
            '구역','개선','불편','진구','시민','코로나','공사장','!!','!!!','부산진','롯데캐슬','횡단보도','철거','신호','복합','위험','개금',
            '부암동','가구','아이파크','자회전','교차로','도시','범천동','삼정','입찰','재생','구청','당감','발생','정비','조치','가로등','공고',
            '교통','동구','마을','방치','보호','전포동','무단','사고']
```

```python
#명사 추출, 단어 벡터화

from konlpy.tag import Komoran
tagger = Komoran()  # 형태소 분석기

def get_nouns(text):
    nouns = tagger.nouns(text)
    return [n for n in nouns if len(n) > 1]

from sklearn.feature_extraction.text import CountVectorizer 

minwon_vectorizer = CountVectorizer(analyzer='word', tokenizer=get_nouns, min_df=5,
                                    stop_words = stopword, ngram_range=(1,2))

tdm = minwon_vectorizer.fit_transform(minlist)
```

```python
tdm.shape #(문서 수, 단어 수)
```

```python
words = minwon_vectorizer.get_feature_names()  #단어 목록 저장
count_minwon = tdm.sum(axis=0) #열별로 모든 행 합계 -> 단어별로 빈도 구함
```

```python
count_minwon.shape
```

```python
count_minwon
```

```python
word_count = list(zip(words, count_minwon.flat)) #단어와 빈도를 튜플 형식으로 매칭
word_count = sorted(word_count, key=lambda x:x[1], reverse=True) 
word_count[:20]  #내림차순 정렬 후 상위 20개만 출력
```

```python
#워드클라우드
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
import nltk 
from nltk.corpus import stopwords

wordcloud = WordCloud(max_font_size=200, font_path='C:/Users/hyunj/AppData/Local/Microsoft/Windows/Fonts/DXMSubtitlesM-KSCpc-EUC-H.ttf',
                background_color='white', width=600, height=800, prefer_horizontal= 1.0)

wordcloud.generate_from_frequencies(dict(word_count)) #빈도 고려하여 워드클라우드 생성


plt.figure(figsize=(8, 6))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```
