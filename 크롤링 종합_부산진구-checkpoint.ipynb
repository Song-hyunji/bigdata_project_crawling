```python
#부산진구 민원 크롤링
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys #저장 시 사용하는 모듈
import pandas as pd
import re

#크롬 드라이버를 사용해서 웹 브라우저 실행
path = "C:/Users/hyunj/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://eminwon.busanjin.go.kr/emwp/gov/mogaha/ntis/web/emwp/cns/action/EmwpCnslWebAction.do?method=selectCnslWebPage&menu_id=EMWPCnslWebInqL&jndinm=EmwpCnslWebEJB&methodnm=selectCnslWebPage&context=NTIS")
time.sleep(1) #인터넷 창이 모두 열릴 때까지 2초 기다림


minwon_df = []
df = pd.DataFrame(minwon_df, columns = ['page','id','title','date'])
num = 0

#셀레니움으로 제목 가져오기
for page in range(40):
    html = driver.page_source  #현재 화면의 html을 모두 불러옴
    soup = BeautifulSoup(html, 'html.parser')  #불러오는 정보를 걸러주는 도구
    
    minwon_dates = [ date.get_text() for date in soup.find_all('td', attrs={'class':'td-date'})] # 기사 작성일
    news_date = []
    for date in minwon_dates:
        if re.search(r'\d+.\d+.\d+.', date) != None: # 기사 작성일 정제
            date = re.sub("-",".",date)
            news_date.append(date)
            
    minwonlist = soup.select("table > tbody > tr > td.td-list > a")
    
    n=-1
    for i in minwonlist:
        n=n+1
        num = num+1
        temp = i.text.strip()
        df = df.append({'page':page+1,'id':num,'date':news_date[n], 'title':temp}, ignore_index=True)
     
    if page > 30:
        element = driver.find_element_by_xpath('/html/body/main/div/form/div[2]/a[11]')
    else: element = driver.find_element_by_xpath('/html/body/main/div/form/div[2]/a[12]')
    driver.execute_script("arguments[0].click();", element)

    
minwon_df = df   
```

```python
#뉴스기사
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

navernews_df = pd.DataFrame()

for page in range(20):
    response = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%B6%80%EC%82%B0%EC%A7%84%EA%B5%AC%20%2B%EB%AC%B8%ED%99%94%20%2B%EC%8B%9C%EC%84%A4%20-%EC%BD%94%EB%A1%9C%EB%82%98%20%2B%EB%B6%80%EC%A1%B1&sort=1&photo=0&field=0&pd=0&ds=&de=&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:dd,p:all,a:all&start={}'.format(page*10+1))
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    news_title = [title['title'] for title in soup.find_all('a', attrs={'class':'news_tit'})] # 기사 제목
    news_url = [ url['href'] for url in soup.find_all('a', attrs={'class':'news_tit'}) ] # 기사 url
    dates = [ date.get_text() for date in soup.find_all('span', attrs={'class':'info'})] # 기사 작성일
    news_date = []
    for date in dates:
        if re.search(r'\d+.\d+.\d+.', date) != None: # 기사 작성일 정제
            news_date.append(date)

    df = pd.DataFrame({'기사작성일':news_date,'기사제목':news_title,'기사주소':news_url})
    navernews_df = pd.concat([navernews_df, df], ignore_index=True)
```

```python
#민원과 뉴스 데이터프레임 index 통일시키기
news_df = navernews_df[['기사제목','기사작성일']]
news_df = news_df.rename(columns={'기사제목':'title','기사작성일':'date'})

#이후 파싱 및 빈도분석에 편하기 위해 전체 결과 csv로 저장
all_df = pd.concat([minwon_df[['title','date']], news_df], ignore_index=True)
all_df.to_csv('부산진구 종합 크롤링.csv', mode="w")
```

```python
#파싱
from konlpy.tag import Twitter 
from collections import Counter

#형태소 나누기
twitter = Twitter() 
morphs = [] 

for sentence in totallist: 
    morphs.append(twitter.pos(sentence)) 
    #print(morphs)

#명사, 형용사, 부사만 골라내기
noun_adj_adv_list=[] 
for sentence in morphs :
    for word, tag in sentence :
        if tag in ['Noun'] and ("것" not in word) and ("내" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word): 
            noun_adj_adv_list.append(word) 
            
#빈도 추출
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
```
