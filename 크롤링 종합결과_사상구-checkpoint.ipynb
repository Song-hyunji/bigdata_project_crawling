```python
#사상구 민원 크롤링
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import re
import sys #저장 시 사용하는 모듈
import pandas as pd

#크롬 드라이버를 사용해서 웹 브라우저 실행
path = "C:/Users/hyunj/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://eminwon.sasang.go.kr/emwp/gov/mogaha/ntis/web/emwp/cns/action/EmwpCnslWebAction.do?method=selectCnslWebPage&menu_id=EMWPCnslWebInqL&jndinm=EmwpCnslWebEJB&methodnm=selectCnslWebPage&context=NTIS")
time.sleep(1) #인터넷 창이 모두 열릴 때까지 2초 기다림

minwon_df = []
df = pd.DataFrame(minwon_df, columns = ['page','id','title','date'])
num = 0

#셀레니움으로 제목 가져오기
for page in range(40):
    html = driver.page_source  #현재 화면의 html을 모두 불러옴
    soup = BeautifulSoup(html, 'html.parser')  #불러오는 정보를 걸러주는 도구
    
    minwon_dates = [ date.get_text() for date in soup.find_all('td', attrs={'class':'td-date'})] # 기사 작성일
    news_date = []
    for date in minwon_dates:
        if re.search(r'\d+.\d+.\d+.', date) != None: # 기사 작성일 정제
            date = re.sub("-",".",date)
            news_date.append(date)
            
    minwonlist = soup.select("table > tbody > tr > td.td-list > a")
    n=-1
    for i in minwonlist:
        n=n+1
        num = num+1
        temp = i.text.strip()
        if '르네시떼' in temp:
            continue
        df = df.append({'page':page+1,'id':num,'date':news_date[n], 'title':temp}, ignore_index=True)

    element = driver.find_element_by_xpath('/html/body/main/div/form/div[2]/a[12]')
    driver.execute_script("arguments[0].click();", element)
    
minwon_df = df
```

```python
#예산편성에 바란다
import urllib3
import requests
import re
from bs4 import BeautifulSoup

urllib3.disable_warnings()

#마지막 페이지 찾기
max_page = 0
response = requests.get('https://www.sasang.go.kr/board/list.sasang?boardId=BBS_0000025&menuCd=DOM_000000102006001001', verify=False)
html = response.text
soup = BeautifulSoup(html, 'html.parser')

for i in soup.find(class_='page').findAll("a"):
    max_page = i.text
max_page = int(max_page)+1
    
    
hope_df = pd.DataFrame()
hope_df = pd.DataFrame(hope_df, columns = ['title','date'])
    
#제목 추출하기
titlelist = []
for page in range(1,max_page):
    response = requests.get('https://www.sasang.go.kr/board/list.sasang?boardId=BBS_0000025&menuCd=DOM_000000102006001001&startPage={}'.format(page), verify=False)

    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    hope_dates = [ date.get_text() for date in soup.find_all('td', attrs={'class':'data'})] #민원작성일
    hopedate = []
    
    ch = "\n\t\r "
    
    for date in hope_dates:
        date = ''.join(x for x in date if x not in ch)
        date = re.sub("-",".",date)
        hopedate.append(date)
            
    suggestlist = soup.select('table > tbody > tr > td.subject')

    n=-1
    for suggest in suggestlist:
        title = suggest.select_one('a')['title']
        titlelist.append(title)
        n=n+1
        hope_df = hope_df.append({'date':hopedate[n],'title':title}, ignore_index=True)
```

```python
hope_df.to_csv('부산사상구 예산편성 민원.csv', encoding='cp949')
```

```python
#뉴스기사
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

navernews_df = pd.DataFrame()

for page in range(20):
    response = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%B6%80%EC%82%B0%20%EC%82%AC%EC%83%81%EA%B5%AC%20%2B%EB%AC%B8%ED%99%94%20%2B%EC%8B%9C%EC%84%A4%20-%EC%BD%94%EB%A1%9C%EB%82%98%20%2B%EB%B6%80%EC%A1%B1&sort=1&photo=0&field=0&pd=0&ds=&de=&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:dd,p:all,a:all&start={}'.format(page*10+1))
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    news_title = [title['title'] for title in soup.find_all('a', attrs={'class':'news_tit'})] # 기사 제목
    news_url = [ url['href'] for url in soup.find_all('a', attrs={'class':'news_tit'}) ] # 기사 url
    dates = [ date.get_text() for date in soup.find_all('span', attrs={'class':'info'})] # 기사 작성일
    news_date = []
    for date in dates:
        if re.search(r'\d+.\d+.\d+.', date) != None: # 기사 작성일 정제
            news_date.append(date)

    df = pd.DataFrame({'기사작성일':news_date,'기사제목':news_title,'기사주소':news_url})
    navernews_df = pd.concat([navernews_df, df], ignore_index=True)
```

```python
#파싱
from konlpy.tag import Twitter 
from collections import Counter

#형태소 나누기
twitter = Twitter() 
morphs = [] 

for sentence in totallist: 
    morphs.append(twitter.pos(sentence)) 
    #print(morphs)

#명사, 형용사, 부사만 골라내기
noun_adj_adv_list=[] 
for sentence in morphs :
    for word, tag in sentence :
        if tag in ['Noun'] and ("것" not in word) and ("내" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word): 
            noun_adj_adv_list.append(word) 
            
#빈도 추출
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
```
